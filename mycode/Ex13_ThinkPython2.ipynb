## Ex 13.5  
Write a function named choose_from_hist that takes a histogram as defined in Section 11.2 and returns a random value from the histogram, chosen with probability in proportion to frequency. For example, for this histogram:
```
>>> t = ['a', 'a', 'b']
>>> hist = histogram(t)
>>> hist
{'a': 2, 'b': 1}
```
your function should return 'a' with probability 2/3 and 'b' with probability 1/3.

````python
def histogram(s):
    d = dict()
    for c in s:
        if c not in d:
            d[c] = 1
        else:
            d[c] += 1
    return d

t = ['a', 'a', 'b']
hist = histogram(t)
````

````python
def choose_from_hist(hist):
    sample = []
    for k, v in hist.items():
#         sample.extend([k for n in range(v)])
        sample.extend([k] * v )
    return random.choice(sample)
````

````python
def gen_sample(hist, n):
    c = 0
    sample = []
    for m in range(n):
        l = choose_from_hist(hist)
        sample.append(l)
    
    return histogram(sample)
````

````python
n = 2000
t = ['a', 'a', 'b']
hist = histogram(t)

print(f"Generated histogram for {n} trials")
gen = gen_sample(hist, n)
print(gen)
print(f"{gen['a']/(gen['a']+gen['b']):.2%}")
````

````python

def bisearch_dict(target, dict):
    sz = len(dict)

    keys = list(dict.keys())
    val = lambda n: dict[keys[n]]
    
    if sz == 0:
        return -1
    if sz == 1:
        if target == val(0):
            return 0
        else:
            return -1

    if target < val(0) or target > val(sz-1):
        return -1
    if target == val(sz-1):
        return sz - 1
    
    # Treat as binary search over segments not points
    s = 0
    e = sz - 2
    
    while e >= s:
        m = (s + e)//2
        if val(m) <= target < val(m+1):
            return m
        if target >= val(m+1):
            s = m + 1
        else:
            e = m - 1
    return -1
````
